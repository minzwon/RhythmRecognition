### Motivation

Music information retrieval (MIR) researchers adopted deep convolutional neural networks (DCNN) in many classification/regression tasks for music and they reported remarkable results. However, when I try visualization techniques, such as Grad-CAM, guided backpropagation, or kernel maximization, it was difficult to find explainable music semantics from there. I am curious if the network learned (or comprehended) music in proper way.

Especially, I wanted to tackle the problem of the rhythm classification. Since, DCNN has fixed size of receptive fields, it's tempo variant. For human music perception, we can perceive rhythms with a same pattern but different tempi as a single rhythmic pattern. So, I wanted to know if conventional DCNNs can learn proper rhythmic patterns in their classification task.

### Extended ballroom dataset

Ballroom dataset, which consists of 698 audio excerpts, is a broadly used dataset for the rhythm classification. Extended ballroom dataset is an extended version of the Ballroom dataset. In this research, I used 9 main genres (3992 audio excerpts) of the Extended ballroom dataset.

As you can see on the boxplot, and as many researchers pointed out before, genre classes in the dataset can be distinguished with the BPM. Also, some genres can be distinguished only by the existence of certain instruments. What DCNN will learn in this condition?

### Preprocessing

I compared two inputs for the DCNN. One is using mel spectrograms as its input and another is using mel spectrograms of transposed audio. I did the preprocessing so that machine can solely concentrate on modeling rhythmic patterns apart from tempo variances. Since time stretching can change the timbre of the audio, I used the transposing method below.

- Beat detection.
- If inter-beat-interval is bigger than 2048 samples, put more beats in the middle of two beats. Here I define 2048 samples as a tatum.
- Crop 2048 samples from each beat onset and concatenate them. I hypothesized that the information around beat onsets are more important than others.
- Get mel spectrograms from the transposed audio.

### Experiment I

First of all, I visualized learned features of the last layer using activation maximization (you can find a good tutorial [here](https://github.com/raghakot/keras-vis/tree/master/examples/vggnet)). Visualized genre kernels are in alphabetic order:

**Chacha Foxtrot Jive **

**Quickstep Rumba Samba**

**Tango Viennesewaltz Waltz**

```python
img3 = cv2.imread('./figures/init_normal.png')
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img4 = cv2.imread('./figures/init_minz.png')
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.imshow(img3)
plt.xticks([])
plt.yticks([])
plt.title('without preprocessing', size=17)
plt.subplot(1,2,2)
plt.imshow(img4)
plt.xticks([])
plt.yticks([])
plt.title('with preprocessing', size=17)
plt.show()
```

In both cases, networks seem to learn temporal patterns. Now I experimented with time-stretched audio to see how much the networks are robust toward tempo variations. The x-axis is time-stretching rate (3%, 5%, 7%, 9%, 11%, 13%, 15%, 17%, 19%) and y-axis is f1-measure of each genre. Genre is in same order as before. A vanilla DCNN was quite robust than I expected. Since _Waltz_ is the slowest genre, it's robust in both networks. But most genres show more robustness with preprocessing as expected. 

F1-measures of _Chacha_ and _Tango_ plunged only with 3% time-stretching. My assumption is that the network learned specific timbre of _Chacha_ and _Tango_ but time-stretching ruined its spectral shape. With this motivation I did the **Experiment II**.

```python
img5 = cv2.imread('./figures/f1_normal.png')
img5 = cv2.cvtColor(img5, cv2.COLOR_BGR2RGB)
img6 = cv2.imread('./figures/f1_minz.png')
img6 = cv2.cvtColor(img6, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.imshow(img5)
plt.xticks([])
plt.yticks([])
plt.title('without preprocessing', size=17)
plt.subplot(1,2,2)
plt.imshow(img6)
plt.xticks([])
plt.yticks([])
plt.title('with preprocessing', size=17)
plt.show()
```
